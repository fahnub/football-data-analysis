```python
##Import Libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
##Load Dataset

df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
##Add More Player Stats Columns (Playtime and Goals/Match)

df['Mins_Per_Match'] = (df['Mins'] / df['Matches']).astype(int)
df['Goals_Per_Match'] = (df['Goals'] / df['Matches']).astype(float)
df.head()
```

```python
##Total Goals Scored in the 2020-21 Season

Total_Goals = df['Goals'].sum()
Total_Goals
```

```python
##Penalties Scored

Scored_Penalties = df['Penalty_Goals'].sum()
Scored_Penalties
```

```python
##Penalties Attempted

Penalty_Attempts = df['Penalty_Attempted'].sum()
Penalty_Attempts
```

```python
##Penalties Scored vs Missed

plt.figure(figsize=(13, 6))
Missed_Penalties = df['Penalty_Attempted'].sum() - Scored_Penalties
data = [Missed_Penalties, Scored_Penalties]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
##Unique Playing Positions

df['Position'].unique()
```

```python
##Total Forwards in EPL

df[df['Position'] == 'FW']
```

```python
##Players from differnt Nations

np.size(df['Nationality'].unique())
```

```python
##Nations with highest player count in EPL

nationality = df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind = 'bar', figsize = (12, 6), color = sns.color_palette("magma"))
```

```python
##Clubs with max number of players

df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
##Clubs with min number of players

df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
##Players based on age group

under20 = df[df['Age'] <= 20]
age20_25 = df[(df['Age'] > 20) & (df['Age'] <= 25)]
age25_30 = df[(df['Age'] > 25) & (df['Age'] <= 30)]
above30 = df[df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
age_labels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Player Age Groups', fontsize=20)
plt.pie(x, labels=age_labels, autopct="%.1f%%")
plt.show()
```

```python
##No. of Players under age 20 in each club

players_under_20 = df[df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette('cubehelix'))
```

```python
##Under 20 players in ManU

players_under_20[players_under_20['Club'] == 'Manchester United']
```

```python
##Under 20 players in Chelsea

players_under_20[players_under_20['Club'] == 'Chelsea']
```

```python
##Average age of players in each club

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Club', y = 'Age', data = df)
plt.xticks(rotation = 90)
```

```python
num_player = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending=False)
```

```python
##Total Assists by each club

assists_by_club = pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=assists_by_club.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize = 15)
ax.set_ylabel('Assist Count', fontsize = 15)
plt.xticks(rotation = 90)
plt.title('Clubs and their Total Assist Count', fontsize=20)
```

```python
##Top 10 Assist Providers in EPL

top_10_assist_providers = df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
top_10_assist_providers
```

```python
##Total Goals by each club

goals_by_club = pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=goals_by_club.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel('Club', fontsize = 15)
ax.set_ylabel('Assist Count', fontsize = 15)
plt.xticks(rotation = 90)
plt.title('Clubs and their Total Goal Count', fontsize=20)
```

```python
##Who got the Golden Boot

top_10_scorers = df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns = 'Goals')
top_10_scorers
```

```python
##Highest Goals per match by players

players_with_highest_goal_to_match_ratio = df[['Name', 'Goals_Per_Match', 'Matches', 'Goals']].nlargest(n=10, columns='Goals_Per_Match')
players_with_highest_goal_to_match_ratio
```

```python
##Goals with and without assists

plt.figure(figsize=(14,7))
assists = df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals_Without_Assists', 'Goals_With_Assits']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
##Top 10 players with most yellow cards in 2020-21

player_yellow_card_count = df.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most yellow cards')
c=sns.barplot(x=player_yellow_card_count['Name'], y=player_yellow_card_count['Yellow_Cards'], color='yellow')
plt.xlabel('Players')
plt.ylabel('Yellow Card Count')
c.set_xticklabels(c.get_xticklabels())
c
```
